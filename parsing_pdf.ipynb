```python
import pandas as pd
import numpy as np
import re
```

```python
#this is able to parse the tables in PMDA so much better
import tabula
```

```python
pmda_20_url = 'https://www.pmda.go.jp/files/000242574.pdf'
n_pages_20 = 13

pmda_21_url = 'https://www.pmda.go.jp/files/000245772.pdf'
n_pages_21 = 14
```

```python
def extract_pmda(pages, url):
    good_cols = (['Review Category', 'Approval Date', 'No.', 
                  'Brand Name', 'New Approval/Partial Change', 
                  'Active Ingredient', 'Notes'])
    
    pmda = pd.DataFrame(columns=good_cols)
    
    for page in range(pages-1):
        page += 1
        print('Extracting page ' + str(page))
        ph_df = tabula.read_pdf(url, pages=page)[0].dropna(how='all', axis=1)
        col_dict = dict(zip(ph_df.columns, good_cols))
        print(len(col_dict))
        print(col_dict)
        ph_df = ph_df.rename(columns=col_dict)
        pmda = pmda.append(ph_df)
        print(pmda.columns)
    pmda = pmda.dropna(subset=['Approval Date', 'Brand Name'])
    
    return pmda
```

```python
pmda_2020 = extract_pmda(n_pages_20, pmda_20_url)
```

```python
pmda_2021 = extract_pmda(n_pages_21, pmda_21_url)
```

```python
review_key = tabula.read_pdf(pmda_20_url, pages=13)[0].dropna(how='all', axis=1)
review_key['Review Category'] = [x.split("\r")[0] for x in review_key['Review Category']]
review_key = dict(zip(review_key['Review Category'], review_key['Products']))
```

```python
pmda_20_21 = pmda_2020.append(pmda_2021)

pmda_20_21['New Approval/Partial Change'] = ([x.split('\r')[0] for x in 
                                              pmda_20_21['New Approval/Partial Change']])

pmda_20_21['Product Name'] = ([x.split('\r')[0] for x in 
                                              pmda_20_21['Brand Name']])

pmda_20_21['Applicant Company'] = ([x.split('\r(')[-1][:-1] for x in 
                                              pmda_20_21['Brand Name']])

pmda_20_21['Short Name'] = ([x.split(' ')[0] for x in 
                                              pmda_20_21['Brand Name']])

pmda_20_21['Active Ingredient'] = ([x.replace('\r', ' ') for x in 
                                              pmda_20_21['Active Ingredient']])

pmda_20_21['Notes'] = ([x.replace('\r', ' ') for x in 
                                              pmda_20_21['Notes']])

pmda_20_21['Approval Date'] = ([x.replace('.', '') for x in 
                                              pmda_20_21['Approval Date']])

pmda_20_21['Review Category'] = ([str(x).replace('\r', ' ') for x in 
                                              pmda_20_21['Review Category']])

pmda_20_21['Review Category'] = ([str(x).replace('.0', '') for x in 
                                              pmda_20_21['Review Category']])

pmda_20_21['Review Category'] = pmda_20_21['Review Category'].map(review_key)
```

```python
pmda_20_21.to_csv('2020 to 2021 PMDA Approvals.csv', index=False)
```
