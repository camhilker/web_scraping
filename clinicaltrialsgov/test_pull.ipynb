```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv('2023_01_04 Master Clintrials Data.csv')
syn_data = pd.read_csv("2023_03_27_synonyms_ph3.csv")
d_index = pd.read_csv('drug_index.csv')
```

```python
query_mol = [x for x in set(syn_data['molecule_id'])]
```

```python
#getting dates into datetime format for finding min
a = list(data['start_date'])

for i, x in enumerate(a):
    if len(x.split('-')) == 2:
        a[i] = '01-'+x

data['start_date'] = [datetime.strptime(x, '%d-%b-%y') if x != 'no data' else 'no data' for x in a]
```

```python
#this currently prints "log" to console, could have this output a .txt file in the future

def find_min_phase_dates(data, syn_data, d_index, query_mol):
    molecule_list = []
    intervention_list = []
    nct_id_list = []
    title_list = []
    phase_list = []
    min_start_date_list = []
    sponsor_list = []
    
    
    for mol in query_mol:
        syns = syn_data['syn'][syn_data['molecule_id'] == mol]
        syns = [x.lower() for x in syns]
        print('\n\nQuerying', syns, 'for ' + mol)
        
        nct_ids_per_drug = list()
        for query_syn in syns:
            for intervention in d_index['Intervention']:
                if str(query_syn) in str(intervention):
                    intv_ids = list(d_index['NCT ID'][d_index['Intervention'] == intervention].values)
                    nct_ids_per_drug += intv_ids
        print('\n', len(set(nct_ids_per_drug)), 'Trials found for ' + mol)
                
        data_per_drug = data[data['nct_id'].isin(set(nct_ids_per_drug))]
        
        for phase in ['Phase 1', 'Phase 2', 'Phase 3']:
            phase_data_per_drug = data_per_drug[data_per_drug['phase'] == phase]
            
            try:
                min_row_per_phase = (phase_data_per_drug[phase_data_per_drug['start_date'] == 
                                                         phase_data_per_drug['start_date'].min()])
                intervention_list.append(min_row_per_phase['intervention_name'].values[0])
                nct_id_list.append(min_row_per_phase['nct_id'].values[0])
                title_list.append(min_row_per_phase['official_title'].values[0])
                phase_list.append(min_row_per_phase['phase'].values[0])
                min_start_date_list.append(min_row_per_phase['start_date'].values[0])
                sponsor_list.append(min_row_per_phase['sponsors'].values[0])
                molecule_list.append(mol)

                print('\n\nTitle of', phase, 'initial study found:', min_row_per_phase['official_title'].values[0])
                print(phase, 'start date:', min_row_per_phase['start_date'].values[0].strftime('%m/%d/%Y'))
                print('Interventions for study:', min_row_per_phase['intervention_name'].values[0])
                                     
            except:
                print('\n No', phase, 'Data found for', mol)
    
    min_phase_data = pd.DataFrame(columns=['molecule', 'intervention', 'nct_id', 'title', 
                                           'phase', 'min_start_date', 'sponsor'])
    min_phase_data['molecule'] = molecule_list
    min_phase_data['intervention'] = intervention_list
    min_phase_data['nct_id'] = nct_id_list
    min_phase_data['title'] = title_list
    min_phase_data['phase'] = phase_list
    min_phase_data['min_start_date'] = min_start_date_list
    min_phase_data['sponsor'] = sponsor_list
    
    return min_phase_data
```

```python
test_pull = find_min_phase_dates(data, syn_data, d_index, query_mol)
```
